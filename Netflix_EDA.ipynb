# 1. Introduction:






Netflix is a streaming service that offers a wide variety of award-winning TV shows, movies, anime, documentaries and more- on thousands of internet connected devices.


A dataset named `Netflix Movies and TV shows` downloaded from kaggle. A short description regarding dataset is given below-

URL- `https://www.kaggle.com/datasets/muhammadtahir194/netflix-movies-and-tv-shows-dataset`

CSV file size - 3320 KB

Content- 8807 rows, 12 columns

Type of content- TV shows and Movies

Duration range- 1925 to 2021


### Columns Description- 

`show_id`: Unique identifier for each Netflix title. 

`type` : Indicates whether the content is a Movie or a TV Show.

`title` : Name of the movie or TV show.

`director` : Name of the director(s), Some values are missing.

`cast` : List of actors/actresses in the content. Some values are missing.

`country` : The country where the movie/TV show was produced.

`date_added` : The date when the content was added to Netflix.

`release_year`: The year when the content was originally released.

`rating`: Content rating (e.g., TV-MA, TV-14)

`duration`: For Movies: Duration is given in minutes.
            For TV Shows: Number of seasons is provide
            
`listed_in`: Tv show and movies is categorized in one of the following - Dramas, International Movies, Documentaries, stand-Up Comedy Comedies, Dramas, International Movies, Dramas, Independent Movies.

`Description`: describe a overview of content.

# 3. Data cleaning and Preprocessing:



Before Commencing any analysis , it's good practice to do data cleaning and make some preprocessing. First of all, let's see if is there any missing values are there in out dataset. For that, we will use `isna()` method and apply `sum()` to count missing values.

```python
Netflix_df.isna().sum()
```

Oops, here many missing values present. Let's see one by one. First, we will see 10 missing values of "date_added" column by `isnull()` method.

```python
Netflix_df[Netflix_df["date_added"].isnull()]
```

Now we will extract different parts of date i.e, day, month, year, weekday from "date_added" column by using `DatetimeIndex` class.

```python
#Return particular day of date 
Netflix_df["day_added"]=pd.DatetimeIndex(Netflix_df.date_added).day
#Return particular month of date
Netflix_df["month_added"]=pd.DatetimeIndex(Netflix_df.date_added).month
# Return particular year of date
Netflix_df["year_added"]=pd.DatetimeIndex(Netflix_df.date_added).year
```

```python
# let's see our updated dataframe
Netflix_df
```

```python
#extracting notnull year_added and release_year for purpose of their average. This average we will fill in null values of year_added
Notnull_date_added_df = Netflix_df.loc[Netflix_df['year_added'].notnull(), ['year_added','release_year']]
```

```python
Notnull_date_added_df
```

```python
Notnull_date_added_df.isna().sum()
```

```python
#let's get difference of year added and realease year column and stored in Difference column 
Notnull_date_added_df["Difference"]=Notnull_date_added_df["year_added"]-Notnull_date_added_df["release_year"]
```

```python
Notnull_date_added_df
```

```python
Notnull_date_added_df["Difference"].mean()
```

```python
Netflix_df
```

Average difference between duration of release year and added year is 4.68. So, we can assume on an average gap of 5 years. We will assign 5 to missing values of year added of content (netflix movies and tv shows). 

For our further analysis we need to fill missing values of day_added, month_added and year_added. Assuming content addition date is 1st Jan. And added year is 5 year after release year. 
For this we will use `fillna()` method.

```python
Netflix_df["day_added"] = Netflix_df["day_added"].fillna(1)
```

```python
Netflix_df["month_added"] = Netflix_df["month_added"].fillna(1)
```

```python
Netflix_df["year_added"] = Netflix_df["year_added"].fillna(Netflix_df["release_year"]+5)
```

```python
# Now, we will check the rows corresponding to missing values of "date_added" column.  
Netflix_df[Netflix_df["date_added"].isnull()]
```

# 4. Exploratory data analysis (EDA):

## Overview

Exploratory Data Analysis (EDA) examines and visualizes data to understand its main characteristics, identify patterns, spot anomalies, and test hypotheses. It helps summarize the data and uncover insights before applying more advanced data analysis techniques.



Exploratory Data Analysis is a critical step in the data science process. It is the foundation for understanding and interpreting complex data sets. EDA helps data scientists identify patterns, spot anomalies, test hypotheses, and check assumptions through various statistical and graphical techniques.

Exploratory Data Analysis allows the identification of data quality issues, such as missing values or errors, which can be addressed before proceeding to more advanced analysis.

We can get info which director's content people like to watch. for that we will extract directors list and their frequency of watch content. We will use `value_counts()` method of pandas to get series of unique directors and frequency of watching their content.

```python
directors = Netflix_df["director"].value_counts(ascending=False)
```

```python
directors
```

```python
directors_df = Netflix_df["director"].value_counts(ascending=False).rename_axis('direcors_name').reset_index(name='Number of Content watched').head(5)
```

Droping Unknow directors because considering Unknown to analysis doesn't make much sense.

```python
directors_df= directors_df.drop([0])
```

Let's plot a bar chart between director and No. of content watched.
For that we will import powerful and versatile python library `matplotlib.pyplot` , aliased by plt.


```python
import matplotlib.pyplot as plt

x = directors_df["direcors_name"]
y = directors_df["Number of Content watched"]

plt.bar(x,y ,label= "directors", color= "blue", width= 0.5)

plt.xlabel("Directors")
plt.ylabel("No. of content watched")
plt.title("Most watched director's content")
plt.legend()
plt.show()
```

```python
release_year_df=pd.DataFrame(Netflix_df["year"].value_counts()).reset_index()
```

```python
x= release_year_df["year"]
y= release_year_df["count"]
plt.plot(x,y,color="blue", marker=".")
for i in range(len(x)):
    plt.annotate(y[i], (x[i], y[i]), textcoords= "offset points",xytext=(0,3), ha="center")

plt.title("year vs count")
plt.xlabel("Release_year")
plt.ylabel("count")
plt.grid(True)
plt.show()
```

`Insight 3` : From above plot, it can concluded that most of the tv shows and movies are added in year 2019. Total 2016 movies or tv shows added.
